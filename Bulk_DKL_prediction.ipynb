## Predict on E

```python
# =============================================================================
# Section 1: Install and Import Packages
# =============================================================================

# Install gpax if not already installed

import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import gpax
import jax.numpy as jnp

# Set a plotting style based on available styles
available_styles = plt.style.available
print("Available styles:", available_styles)
if 'seaborn' in available_styles:
    plt.style.use('seaborn')
else:
    plt.style.use('default')

# =============================================================================
# Section 2: Define Helper Functions
# =============================================================================

def process_file(file_name, depth_factor, load_factor, prefix, num_segments=5):
    """
    Reads an Excel file with interleaved depth and load columns,
    applies scaling, segments the data over a range of maximum depths,
    and returns a dictionary with segmented series.
    """
    df = pd.read_excel(file_name, header=None)

    # Extract depths (even columns) and loads (odd columns)
    depths = df.iloc[:, ::2].apply(pd.to_numeric, errors='coerce')
    loads  = df.iloc[:, 1::2].apply(pd.to_numeric, errors='coerce')

    # Apply scaling factors
    depths *= depth_factor
    loads  *= load_factor

    data_dict = {}
    for i in range(depths.shape[1]):
        depth_col = depths.iloc[:, i].dropna()
        load_col  = loads.iloc[:, i].dropna()
        for max_depth in range(200, 1900, 100):
            mask = (depth_col >= 0) & (depth_col <= max_depth)
            depth_selected = depth_col[mask]
            load_selected  = load_col[mask]
            if not depth_selected.empty:
                segment_length = max_depth / num_segments
                for segment in range(1, num_segments+1):
                    segment_mask = (depth_selected >= (segment-1)*segment_length) & (depth_selected < segment*segment_length)
                    depth_segment = depth_selected[segment_mask]
                    load_segment  = load_selected[segment_mask]
                    if not depth_segment.empty:
                        depth_key = f"{prefix}_Depth_{max_depth}_Segment_{segment}_{i}"
                        load_key  = f"{prefix}_Load_{max_depth}_Segment_{segment}_{i}"
                        data_dict[depth_key] = depth_segment.reset_index(drop=True)
                        data_dict[load_key]  = load_segment.reset_index(drop=True)
    return data_dict

def polynomial(x, *coeffs, exponents):
    """
    Generic polynomial function defined by the given exponents.
    """
    return sum(c * x**exp for c, exp in zip(coeffs, exponents))

def fit_polynomial(data_dict, exponents):
    """
    Fits a polynomial (with provided exponents) to each segmented depth curve.
    Returns a dictionary mapping each 'Depth' key to its fitted coefficients.
    Ensures depth and load data have the same length for curve fitting.
    """
    results_dict = {}
    for key in data_dict:
        if 'Depth' in key:
            depth_key = key
            load_key  = key.replace('Depth', 'Load')

            # Align depth and load series based on index after dropping NaNs
            aligned_df = pd.DataFrame({'Depth': data_dict[depth_key], 'Load': data_dict[load_key]}).dropna()

            if not aligned_df.empty:
                X = aligned_df['Depth'].values.flatten()
                y = aligned_df['Load'].values.flatten()

                try:
                    popt, _ = curve_fit(lambda x, *params: polynomial(x, *params, exponents=exponents),
                                        X, y, p0=[1]*len(exponents))
                    results_dict[depth_key] = popt
                except RuntimeError:
                    print(f"Fit failed for dataset {key}")
            else:
                print(f"No valid data points for fitting in dataset {key}")

    return results_dict

def get_dkl_func(targets, X, num_segments, exponents):
    """
    Sets up and trains a DKL model with a one-dimensional output.
    """
    rng_key, rng_key_predict = gpax.utils.get_keys()
    input_dim = len(exponents) * num_segments
    dkl = gpax.viDKL(input_dim, 1, 'RBF')
    print(f"Fitting DKL with input_dim={input_dim}, targets_shape={targets.shape}, X_shape={X.shape}")
    dkl.fit(rng_key, X, targets)
    return dkl, rng_key, rng_key_predict

# Fix the UnboundLocalError in split_in_batches
def split_in_batches_fixed(X_new, batch_size, dim=0):
    """Splits array `X_new` into batches along specified dimension `dim`."""
    num_batches = X_new.shape[dim] // batch_size
    X_split = []
    i = -1 # Initialize i before the loop
    for i in range(num_batches):
        if dim == 0:
            X_i = X_new[i*batch_size:(i+1)*batch_size]
        else:
            X_i = X_new[:, i*batch_size:(i+1)*batch_size]
        X_split.append(X_i)
    # Add remaining elements if any
    # Ensure i is defined even if num_batches is 0
    if num_batches == 0:
        i = -1
    X_i = X_new[(i+1)*batch_size:] if dim == 0 else X_new[:, (i+1)*batch_size:]
    if X_i.shape[dim] > 0:
        X_split.append(X_i)
    return X_split

# Replace the original split_in_batches with the fixed one in gpax.utils
gpax.utils.split_in_batches = split_in_batches_fixed


def process_prediction_file(pred_file, prefix, depth_factor, load_factor, num_segments, max_depth_use, exponents):
    """
    Processes a prediction Excel file and returns a feature matrix based on polynomial fits.
    """
    data_pred = process_file(pred_file, depth_factor, load_factor, prefix=prefix, num_segments=num_segments)
    fit_results_pred = fit_polynomial(data_pred, exponents)
    pred_keys = {key: val for key, val in fit_results_pred.items() if f"Depth_{max_depth_use}_" in key}
    segment_keys_by_sample_pred = {}
    for key in pred_keys:
        sample_index = int(key.split('_')[-1])
        segment_keys_by_sample_pred.setdefault(sample_index, []).append(key)
    X_pred = []
    for sample_index in sorted(segment_keys_by_sample_pred.keys()):
        combined_params = []
        for key in sorted(segment_keys_by_sample_pred[sample_index]):
            combined_params.extend(pred_keys[key])
        X_pred.append(combined_params)
    return np.array(X_pred)

def train_and_predict(target, X_train, X_pred, property_name, num_segments, exponents):
    """
    Trains a DKL model for one target property and predicts on the provided feature matrix.
    """
    dkl, rng_key, rng_key_predict = get_dkl_func(target, X_train, num_segments, exponents)
    # Ensure the corrected split_in_batches is used by the DKL model
    import gpax.utils
    gpax.utils.split_in_batches = split_in_batches_fixed
    predictions = dkl.predict_in_batches(rng_key_predict, X_pred, batch_size=3)
    print(f"Predictions for {property_name}:")
    print(predictions)
    return predictions

def plot_predictions(pred_array, title):
    """
    Expects pred_array to be a tuple: (mean predictions, uncertainty estimates)
    and plots the mean prediction for each sample.
    """
    means, uncert = pred_array
    plt.figure(figsize=(6,4))
    plt.plot(means, 'o-', label='Predicted Mean')
    plt.fill_between(np.arange(len(means)),
                     means - uncert,
                     means + uncert, color='gray', alpha=0.3, label='Uncertainty')
    plt.title(title)
    plt.xlabel("Sample Index")
    plt.ylabel("Predicted E")
    plt.legend()
    plt.show()

# =============================================================================
# Section 3: Define File Names and Parameters
# =============================================================================
# Training files
train_curve_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 dis and load 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'
train_props_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 E sigma_y sigma_s 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'

# Prediction (experimental) files
pred_file_tungsten = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Predicted_1_analyzed_depth2.xlsx'
pred_file_mo       = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'
pred_file_copper   = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of 4um Test 2 Sample 5.xlsx'
pred_file_simu     = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'  # Additional prediction file

# Processing parameters
depth_factor  = 1
load_factor   = 1
num_segments  = 1          # Using one segment for simplicity
max_depth_use = 300        # Must match one of the max_depth values used in process_file
exponents     = [1, 2]     # Polynomial exponents

# =============================================================================
# Section 4: Load Training Material Property (E) and Plot Its Distribution
# =============================================================================

df_props = pd.read_excel(train_props_file)
# Assume columns: 0 -> E, 1 -> σᵧ, 2 -> σₛ
target_E = df_props.iloc[:, 0].values  # Use only E (Young's modulus)

print("Training target shape for E:", target_E.shape)

plt.figure(figsize=(6,4))
plt.hist(target_E, bins=20, color='skyblue')
plt.title("Distribution of E (Young's Modulus)")
plt.xlabel("E")
plt.ylabel("Frequency")
plt.show()

# =============================================================================
# Section 5: Process Training Curves and Extract Features for E
# =============================================================================

data_train = process_file(train_curve_file, depth_factor, load_factor, prefix='train', num_segments=num_segments)
fit_results_train = fit_polynomial(data_train, exponents)
# Filter keys for the chosen max_depth (e.g., "Depth_300")
train_keys = {key: val for key, val in fit_results_train.items() if f"Depth_{max_depth_use}_" in key}

# Group coefficients by sample index (assumed to be the integer after the last underscore)
segment_keys_by_sample = {}
for key in train_keys:
    sample_index = int(key.split('_')[-1])
    segment_keys_by_sample.setdefault(sample_index, []).append(key)

X_train = []
for sample_index in sorted(segment_keys_by_sample.keys()):
    combined_params = []
    for key in sorted(segment_keys_by_sample[sample_index]):
        combined_params.extend(train_keys[key])
    X_train.append(combined_params)
X_train = np.array(X_train)
print("Training feature matrix shape:", X_train.shape)

# Optional: Plot a sample curve and its polynomial fit for visual inspection
sample_key = list(train_keys.keys())[0]
depth_series = data_train[sample_key]
load_series  = data_train[sample_key.replace('Depth', 'Load')]
coeffs_sample = train_keys[sample_key]
x_vals = np.linspace(depth_series.min(), depth_series.max(), 100)
y_fit = polynomial(x_vals, *coeffs_sample, exponents=exponents)

plt.figure(figsize=(6,4))
plt.plot(depth_series, load_series, 'o', label='Data')
plt.plot(x_vals, y_fit, '-', label='Polynomial Fit')
plt.title(f"Sample Curve and Fit\nKey: {sample_key}")
plt.xlabel("Depth")
plt.ylabel("Load")
plt.legend()
plt.show()

# =============================================================================
# Section 6: Process Prediction Files and Extract Feature Matrices for E
# =============================================================================

X_pred_tungsten = process_prediction_file(pred_file_tungsten, prefix='tungsten',
                                            depth_factor=depth_factor, load_factor=load_factor,
                                            num_segments=num_segments, max_depth_use=max_depth_use,
                                            exponents=exponents)
print("Tungsten prediction feature matrix shape:", X_pred_tungsten.shape)

X_pred_mo = process_prediction_file(pred_file_mo, prefix='mo',
                                      depth_factor=depth_factor, load_factor=load_factor,
                                      num_segments=num_segments, max_depth_use=max_depth_use,
                                      exponents=exponents)
print("Mo prediction feature matrix shape:", X_pred_mo.shape)

X_pred_copper = process_prediction_file(pred_file_copper, prefix='copper',
                                          depth_factor=depth_factor, load_factor=load_factor,
                                          num_segments=num_segments, max_depth_use=max_depth_use,
                                          exponents=exponents)
print("Copper prediction feature matrix shape:", X_pred_copper.shape)

X_pred_simu = process_prediction_file(pred_file_simu, prefix='simu',
                                        depth_factor=depth_factor, load_factor=load_factor,
                                        num_segments=num_segments, max_depth_use=max_depth_use,
                                        exponents=exponents)
print("Simulation prediction feature matrix shape:", X_pred_simu.shape)

# =============================================================================
# Section 7: Train DKL Model for E and Make Predictions
# =============================================================================

pred_E_tungsten = train_and_predict(target_E, X_train, X_pred_tungsten, "E (Young's Modulus) - Tungsten", num_segments, exponents)
pred_E_mo       = train_and_predict(target_E, X_train, X_pred_mo,       "E (Young's Modulus) - Mo", num_segments, exponents)
pred_E_copper   = train_and_predict(target_E, X_train, X_pred_copper,   "E (Young's Modulus) - Copper", num_segments, exponents)
pred_E_simu     = train_and_predict(target_E, X_train, X_pred_simu,     "E (Young's Modulus) - Simulation", num_segments, exponents)

# =============================================================================
# Section 8: Plot Predictions for E (Young's Modulus)
# =============================================================================

plot_predictions(pred_E_tungsten, "Predicted E (Young's Modulus) - Tungsten")
plot_predictions(pred_E_simu, "Predicted E (Young's Modulus) - Simulation")
```

```python
# =============================================================================
# Section 1: Install and Import Packages
# =============================================================================

import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import gpax
import jax.numpy as jnp

# =============================================================================
# Section 2: Define Helper Functions
# =============================================================================

def process_file(file_name, depth_factor, load_factor, prefix, num_segments=5):
    df = pd.read_excel(file_name, header=None)
    depths = df.iloc[:, ::2].apply(pd.to_numeric, errors='coerce')
    loads  = df.iloc[:, 1::2].apply(pd.to_numeric, errors='coerce')
    depths *= depth_factor
    loads  *= load_factor

    data_dict = {}
    for i in range(depths.shape[1]):
        depth_col = depths.iloc[:, i].dropna()
        load_col  = loads.iloc[:, i].dropna()
        for max_depth in range(200, 1900, 100):
            mask = (depth_col >= 0) & (depth_col <= max_depth)
            d_sel = depth_col[mask]
            l_sel = load_col[mask]
            if d_sel.empty:
                continue
            segment_length = max_depth / num_segments
            for seg in range(1, num_segments+1):
                seg_mask = (d_sel >= (seg-1)*segment_length) & (d_sel < seg*segment_length)
                d_seg = d_sel[seg_mask]
                l_seg = l_sel[seg_mask]
                if d_seg.empty:
                    continue
                key_d = f"{prefix}_Depth_{max_depth}_Segment_{seg}_{i}"
                key_l = f"{prefix}_Load_{max_depth}_Segment_{seg}_{i}"
                data_dict[key_d] = d_seg.reset_index(drop=True)
                data_dict[key_l] = l_seg.reset_index(drop=True)
    return data_dict


def polynomial(x, *coeffs, exponents):
    return sum(c * x**e for c, e in zip(coeffs, exponents))


def fit_polynomial(data_dict, exponents):
    results = {}
    for key in data_dict:
        if 'Depth' not in key:
            continue
        depth_key = key
        load_key  = key.replace('Depth', 'Load')
        df_aligned = pd.DataFrame({
            'Depth': data_dict[depth_key],
            'Load' : data_dict[load_key]
        }).dropna()
        if df_aligned.empty:
            print(f"No data for {depth_key}")
            continue
        X = df_aligned['Depth'].values
        y = df_aligned['Load'].values
        try:
            popt, _ = curve_fit(
                lambda x, *p: polynomial(x, *p, exponents=exponents),
                X, y, p0=[1]*len(exponents)
            )
            results[depth_key] = popt
        except RuntimeError:
            print(f"Fit failed for {depth_key}")
    return results


def get_dkl_func(targets, X, num_segments, exponents):
    rng_key, rng_key_pred = gpax.utils.get_keys()
    input_dim = len(exponents) * num_segments
    dkl = gpax.viDKL(input_dim, 1, 'RBF')
    print(f"Training DKL: input_dim={input_dim}, X.shape={X.shape}, targets.shape={targets.shape}")
    dkl.fit(rng_key, X, targets)
    return dkl, rng_key, rng_key_pred


def process_prediction_file(pred_file, prefix, depth_factor, load_factor, num_segments, max_depth_use, exponents):
    data_pred = process_file(pred_file, depth_factor, load_factor, prefix, num_segments)
    fit_pred  = fit_polynomial(data_pred, exponents)
    sel = {k: v for k, v in fit_pred.items() if f"Depth_{max_depth_use}_" in k}
    by_sample = {}
    for k in sel:
        idx = int(k.split('_')[-1])
        by_sample.setdefault(idx, []).append(k)
    Xp = []
    for idx in sorted(by_sample):
        coeffs = []
        for k in sorted(by_sample[idx]):
            coeffs.extend(sel[k])
        Xp.append(coeffs)
    return np.array(Xp)


def train_and_predict(targets, X_train, X_pred, prop_name, num_segments, exponents):
    def split_in_batches_fixed(X_new, batch_size, dim=0):
        n = X_new.shape[dim] // batch_size
        chunks = [
            X_new[i*batch_size:(i+1)*batch_size] if dim==0 else X_new[:, i*batch_size:(i+1)*batch_size]
            for i in range(n)
        ]
        rem = X_new[n*batch_size:] if dim==0 else X_new[:, n*batch_size:]
        if rem.shape[dim] > 0:
            chunks.append(rem)
        return chunks

    original = gpax.utils.split_in_batches
    gpax.utils.split_in_batches = split_in_batches_fixed

    dkl, rk, rk_pred = get_dkl_func(targets, X_train, num_segments, exponents)
    preds = dkl.predict_in_batches(rk_pred, X_pred, batch_size=3)

    gpax.utils.split_in_batches = original
    print(f"Predictions for {prop_name}:\n", preds)
    return preds


def plot_predictions(pred_array, title):
    means, unc = pred_array
    plt.figure(figsize=(6,4))
    plt.plot(means, 'o-', label='Mean')
    plt.fill_between(np.arange(len(means)), means-unc, means+unc,
                     color='gray', alpha=0.3, label='Uncertainty')
    plt.title(title)
    plt.xlabel("Sample Index")
    plt.ylabel(title)
    plt.legend()
    plt.show()

# =============================================================================
# Section 3: File Paths & Parameters
# =============================================================================

train_curve_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 dis and load 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'
train_props_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 E sigma_y sigma_s 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'

pred_file_tungsten = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Predicted_1_analyzed_depth2.xlsx'
pred_file_mo       = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'
pred_file_copper   = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of 4um Test 2 Sample 5.xlsx'
pred_file_simu     = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of around 2um Test 1 Round 2 Sample 5.xlsx'

depth_factor  = 1
load_factor   = 1
num_segments  = 1
max_depth_use = 300
exponents     = [1, 2]

# =============================================================================
# Section 4: Load Training Target σₛ and Plot Distribution
# =============================================================================

df_props       = pd.read_excel(train_props_file)
target_sigma_s = df_props.iloc[:, 2].values
print("Training target shape for σₛ:", target_sigma_s.shape)

plt.figure(figsize=(6,4))
plt.hist(target_sigma_s, bins=20)
plt.title("Distribution of σₛ")
plt.xlabel("σₛ")
plt.ylabel("Frequency")
plt.show()

# =============================================================================
# Section 5: Build Training Feature Matrix
# =============================================================================

data_train = process_file(train_curve_file, depth_factor, load_factor, 'train', num_segments)
fit_train  = fit_polynomial(data_train, exponents)
train_sel  = {k:v for k,v in fit_train.items() if f"Depth_{max_depth_use}_" in k}

by_sample = {}
for k in train_sel:
    idx = int(k.split('_')[-1])
    by_sample.setdefault(idx, []).append(k)

X_train = []
for idx in sorted(by_sample):
    coeffs = []
    for k in sorted(by_sample[idx]):
        coeffs.extend(train_sel[k])
    X_train.append(coeffs)
X_train = np.array(X_train)
print("X_train shape:", X_train.shape)

# Optional: plot one fit
sample_key = list(train_sel.keys())[0]
d = data_train[sample_key]
l = data_train[sample_key.replace('Depth','Load')]
c = train_sel[sample_key]
xv = np.linspace(d.min(), d.max(), 100)
yv = polynomial(xv, *c, exponents=exponents)
plt.figure(figsize=(6,4))
plt.plot(d, l, 'o', label='Data')
plt.plot(xv, yv, '-', label='Fit')
plt.title(f"Sample Fit: {sample_key}")
plt.xlabel("Depth")
plt.ylabel("Load")
plt.legend()
plt.show()

# =============================================================================
# Section 6: Build Prediction Feature Matrices
# =============================================================================

pred_files = {
    'tungsten': pred_file_tungsten,
    'mo'      : pred_file_mo,
    'copper'  : pred_file_copper,
    'simu'    : pred_file_simu
}

X_preds = {}
for name, path in pred_files.items():
    Xp = process_prediction_file(
        path, name, depth_factor, load_factor,
        num_segments, max_depth_use, exponents
    )
    print(f"{name} X_pred shape:", Xp.shape)
    X_preds[name] = Xp

# =============================================================================
# Section 7: Train DKL on σₛ and Predict
# =============================================================================

predictions = {}
for name, Xp in X_preds.items():
    pred = train_and_predict(
        target_sigma_s, X_train, Xp,
        f"σₛ - {name.capitalize()}", num_segments, exponents
    )
    predictions[name] = pred

# =============================================================================
# Section 8: Plot σₛ Predictions
# =============================================================================

for name, pred in predictions.items():
    plot_predictions(pred, f"Predicted σₛ - {name.capitalize()}")
```

```python
# =============================================================================
# Section 1: Imports (install gpax separately: pip install git+https://github.com/ziatdinovmax/gpax)
# =============================================================================

import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import gpax
import jax.numpy as jnp

# Set a plotting style. If 'seaborn' is not available, default to 'default'.
available_styles = plt.style.available
print("Available styles:", available_styles)
if 'seaborn' in available_styles:
    plt.style.use('seaborn')
else:
    plt.style.use('default')

# Fix the UnboundLocalError in split_in_batches
def split_in_batches_fixed(X_new, batch_size, dim=0):
    """Splits array `X_new` into batches along specified dimension `dim`."""
    num_batches = X_new.shape[dim] // batch_size
    X_split = []
    i = -1 # Initialize i before the loop
    for i in range(num_batches):
        if dim == 0:
            X_i = X_new[i*batch_size:(i+1)*batch_size]
        else:
            X_i = X_new[:, i*batch_size:(i+1)*batch_size]
        X_split.append(X_i)
    # Add remaining elements if any
    # Ensure i is defined even if num_batches is 0
    if num_batches == 0:
        i = -1
    X_i = X_new[(i+1)*batch_size:] if dim == 0 else X_new[:, (i+1)*batch_size:]
    if X_i.shape[dim] > 0:
        X_split.append(X_i)
    return X_split

# Replace the original split_in_batches with the fixed one in gpax.utils
gpax.utils.split_in_batches = split_in_batches_fixed


# =============================================================================
# Section 2: Helper Functions
# =============================================================================

def process_file(file_name, depth_factor, load_factor, prefix, num_segments=5):
    df = pd.read_excel(file_name, header=None)
    depths = df.iloc[:, ::2].apply(pd.to_numeric, errors='coerce') * depth_factor
    loads  = df.iloc[:, 1::2].apply(pd.to_numeric, errors='coerce') * load_factor
    data_dict = {}
    for i in range(depths.shape[1]):
        d_col = depths.iloc[:, i].dropna()
        l_col = loads.iloc[:, i].dropna()
        for max_depth in range(200, 1900, 100):
            mask = (d_col >= 0) & (d_col <= max_depth)
            d_sel, l_sel = d_col[mask], l_col[mask]
            if not d_sel.empty:
                seg_len = max_depth / num_segments
                for seg in range(1, num_segments+1):
                    seg_mask = (d_sel >= (seg-1)*seg_len) & (d_sel < seg*seg_len)
                    d_seg, l_seg = d_sel[seg_mask], l_sel[seg_mask]
                    if not d_seg.empty:
                        d_key = f"{prefix}_Depth_{max_depth}_Segment_{seg}_{i}"
                        l_key = f"{prefix}_Load_{max_depth}_Segment_{seg}_{i}"
                        data_dict[d_key] = d_seg.reset_index(drop=True)
                        data_dict[l_key] = l_seg.reset_index(drop=True)
    return data_dict


def polynomial(x, *coeffs, exponents):
    return sum(c * x**e for c, e in zip(coeffs, exponents))


def fit_polynomial(data_dict, exponents):
    """
    Fits a polynomial (with provided exponents) to each segmented depth curve.
    Returns a dictionary mapping each 'Depth' key to its fitted coefficients.
    Ensures depth and load data have the same length for curve fitting.
    """
    results = {}
    for d_key, series in data_dict.items():
        if 'Depth' in d_key:
            l_key = d_key.replace('Depth', 'Load')

            # Align depth and load series based on index after dropping NaNs
            aligned_df = pd.DataFrame({'Depth': data_dict[d_key], 'Load': data_dict[l_key]}).dropna()

            if not aligned_df.empty:
                X = aligned_df['Depth'].values.flatten()
                y = aligned_df['Load'].values.flatten()

                try:
                    popt, _ = curve_fit(
                        lambda x, *p: polynomial(x, *p, exponents=exponents),
                        X, y,
                        p0=[1]*len(exponents)
                    )
                    results[d_key] = popt
                except RuntimeError:
                    print(f"Fit failed for {d_key}")
            else:
                print(f"No valid data points for fitting in dataset {d_key}")

    return results


def get_dkl_func(targets, X, num_segments, exponents):
    rng_train, rng_pred = gpax.utils.get_keys()
    input_dim = len(exponents) * num_segments
    dkl = gpax.viDKL(input_dim, 1, 'RBF')
    print(f"[DKL] input_dim={input_dim}, X={X.shape}, y={targets.shape}")
    dkl.fit(rng_train, X, targets)
    return dkl, rng_train, rng_pred


def process_prediction_file(pred_file, prefix,
                            depth_factor, load_factor,
                            num_segments, max_depth_use, exponents):
    data_pred = process_file(pred_file, depth_factor, load_factor, prefix, num_segments)
    fit_pred  = fit_polynomial(data_pred, exponents)
    sel = {k: v for k, v in fit_pred.items() if f"Depth_{max_depth_use}_" in k}

    grouped = {}
    for k in sel:
        idx = int(k.split('_')[-1])
        grouped.setdefault(idx, []).append(k)

    X_pred = []
    for idx in sorted(grouped):
        row = []
        for k in sorted(grouped[idx]):
            row.extend(sel[k])
        X_pred.append(row)

    return np.array(X_pred)


def train_and_predict(log_target, X_train, X_pred,
                      label, num_segments, exponents):
    # Train on log-space targets
    dkl, rng_train, rng_pred = get_dkl_func(log_target, X_train, num_segments, exponents)
    mean_log, uncert_log = dkl.predict_in_batches(rng_pred, X_pred, batch_size=3)

    # Back-transform to positive space
    mean   = np.exp(mean_log)
    uncert = mean * uncert_log  # delta-method

    print(f"{label} → mean: {mean}\n            uncert: {uncert}")
    return mean, uncert


def plot_predictions(pred, title):
    mean, uncert = pred
    plt.figure(figsize=(6,4))
    plt.plot(mean, 'o-', label='Predicted Mean')
    plt.fill_between(
        np.arange(len(mean)),
        mean-uncert,
        mean+uncert,
        alpha=0.3, label='Uncertainty'
    )
    plt.title(title)
    plt.xlabel("Sample Index")
    plt.ylabel("σₛ (MPa)")
    plt.legend()
    plt.show()


# =============================================================================
# Section 3: File Paths & Parameters
# =============================================================================
# Training files
train_curve_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 dis and load 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'
train_props_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 E sigma_y sigma_s 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'

pred_files = {
    'Tungsten': '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of around 2um Test 1 Round 2 Sample 5.xlsx',
    'Mo':        '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Predicted_1_analyzed_depth2.xlsx',
    'Copper':    '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of 4um Test 2 Sample 5.xlsx',
    'Simu':      '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Copper Experiment Thickness of around 2um Test 1 Round 2 Sample 5.xlsx'  # Additional prediction file
}

depth_factor  = 1
load_factor   = 1
num_segments  = 1
max_depth_use = 300
exponents     = [1, 2]


# =============================================================================
# Section 4: Load & Log-Transform σₛ
# =============================================================================

df_props     = pd.read_excel(train_props_file)
raw_sigma_s  = df_props.iloc[:, 2].values
log_sigma_s  = np.log(raw_sigma_s)

plt.figure(figsize=(6,4))
plt.hist(raw_sigma_s, bins=20)
plt.title("σₛ Distribution")
plt.xlabel("σₛ"); plt.ylabel("Count")
plt.show()


# =============================================================================
# Section 5: Build Training Feature Matrix
# =============================================================================

data_train = process_file(train_curve_file, depth_factor, load_factor, 'train', num_segments)
fit_train  = fit_polynomial(data_train, exponents)
train_sel  = {k: v for k, v in fit_train.items() if f"Depth_{max_depth_use}_" in k}

groups = {}
for k in train_sel:
    idx = int(k.split('_')[-1])
    groups.setdefault(idx, []).append(k)

X_train = []
for idx in sorted(groups):
    row = []
    for k in sorted(groups[idx]):
        row.extend(train_sel[k])
    X_train.append(row)
X_train = np.array(X_train)

print("X_train shape:", X_train.shape)


# =============================================================================
# Section 6: Process Prediction Sets & Predict
# =============================================================================

pred_results = {}
for name, path in pred_files.items():
    Xp = process_prediction_file(path, name.lower(),
                                 depth_factor, load_factor,
                                 num_segments, max_depth_use, exponents)
    print(f"{name} X_pred shape:", Xp.shape)
    pred_results[name] = train_and_predict(
        log_sigma_s, X_train, Xp,
        f"σₛ (Ultimate Strength) - {name}",
        num_segments, exponents
    )


# =============================================================================
# Section 7: Plot All Predictions
# =============================================================================

for name, pred in pred_results.items():
    plot_predictions(pred, f"Predicted σₛ - {name}")
```

```python
# =============================================================================
# Section 1: Install and Import Packages
# =============================================================================

import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import gpax
import jax.numpy as jnp
from sklearn.preprocessing import StandardScaler

# =============================================================================
# Section 2: Define Helper Functions
# =============================================================================

def process_file(file_name, depth_factor, load_factor, prefix, num_segments=5):
    df = pd.read_excel(file_name, header=None)
    depths = df.iloc[:, ::2].apply(pd.to_numeric, errors='coerce')
    loads  = df.iloc[:, 1::2].apply(pd.to_numeric, errors='coerce')
    depths *= depth_factor
    loads  *= load_factor

    data_dict = {}
    for i in range(depths.shape[1]):
        depth_col = depths.iloc[:, i].dropna()
        load_col  = loads.iloc[:, i].dropna()
        for max_depth in range(200, 1900, 100):
            mask = (depth_col >= 0) & (depth_col <= max_depth)
            d_sel = depth_col[mask]
            l_sel = load_col[mask]
            if d_sel.empty:
                continue
            segment_length = max_depth / num_segments
            for seg in range(1, num_segments+1):
                seg_mask = (d_sel >= (seg-1)*segment_length) & (d_sel < seg*segment_length)
                d_seg = d_sel[seg_mask]
                l_seg = l_sel[seg_mask]
                if d_seg.empty:
                    continue
                key_d = f"{prefix}_Depth_{max_depth}_Segment_{seg}_{i}"
                key_l = f"{prefix}_Load_{max_depth}_Segment_{seg}_{i}"
                data_dict[key_d] = d_seg.reset_index(drop=True)
                data_dict[key_l] = l_seg.reset_index(drop=True)
    return data_dict


def polynomial(x, *coeffs, exponents):
    return sum(c * x**e for c, e in zip(coeffs, exponents))


def fit_polynomial(data_dict, exponents):
    results = {}
    for key in data_dict:
        if 'Depth' not in key:
            continue
        depth_key = key
        load_key  = key.replace('Depth', 'Load')
        df_aligned = pd.DataFrame({
            'Depth': data_dict[depth_key],
            'Load' : data_dict[load_key]
        }).dropna()
        if df_aligned.empty:
            print(f"No data for {depth_key}")
            continue
        X = df_aligned['Depth'].values
        y = df_aligned['Load'].values
        try:
            popt, _ = curve_fit(
                lambda x, *p: polynomial(x, *p, exponents=exponents),
                X, y, p0=[1]*len(exponents)
            )
            results[depth_key] = popt
        except RuntimeError:
            print(f"Fit failed for {depth_key}")
    return results


def get_dkl_func(targets, X, num_segments, exponents):
    rng_key, rng_key_pred = gpax.utils.get_keys()
    input_dim = len(exponents) * num_segments
    dkl = gpax.viDKL(input_dim, 1, 'RBF')
    print(f"Training DKL: input_dim={input_dim}, X.shape={X.shape}, targets.shape={targets.shape}")
    dkl.fit(rng_key, X, targets)
    return dkl, rng_key, rng_key_pred


def process_prediction_file(pred_file, prefix, depth_factor, load_factor, num_segments, max_depth_use, exponents):
    data_pred = process_file(pred_file, depth_factor, load_factor, prefix, num_segments)
    fit_pred  = fit_polynomial(data_pred, exponents)
    sel = {k: v for k, v in fit_pred.items() if f"Depth_{max_depth_use}_" in k}
    by_sample = {}
    for k in sel:
        idx = int(k.split('_')[-1])
        by_sample.setdefault(idx, []).append(k)
    Xp = []
    for idx in sorted(by_sample):
        coeffs = []
        for k in sorted(by_sample[idx]):
            coeffs.extend(sel[k])
        Xp.append(coeffs)
    return np.array(Xp)


def train_and_predict(targets, X_train, X_pred, prop_name, num_segments, exponents, scaler_y=None):
    def split_in_batches_fixed(X_new, batch_size, dim=0):
        n = X_new.shape[dim] // batch_size
        chunks = [
            X_new[i*batch_size:(i+1)*batch_size] if dim==0 else X_new[:, i*batch_size:(i+1)*batch_size]
            for i in range(n)
        ]
        rem = X_new[n*batch_size:] if dim==0 else X_new[:, n*batch_size:]
        if rem.shape[dim] > 0:
            chunks.append(rem)
        return chunks

    original = gpax.utils.split_in_batches
    gpax.utils.split_in_batches = split_in_batches_fixed

    dkl, rk, rk_pred = get_dkl_func(targets, X_train, num_segments, exponents)
    means, unc = dkl.predict_in_batches(rk_pred, X_pred, batch_size=3)

    gpax.utils.split_in_batches = original

    # Inverse-scale predictions if scaler_y provided
    if scaler_y is not None:
        means = scaler_y.inverse_transform(means.reshape(-1,1)).flatten()
        unc   = unc * scaler_y.scale_[0]
    print(f"Predictions for {prop_name}:\nMeans: {means}\nUncertainty: {unc}")
    return means, unc


def plot_predictions(means, unc, title):
    plt.figure(figsize=(6,4))
    plt.plot(means, 'o-', label='Mean')
    plt.fill_between(np.arange(len(means)), means-unc, means+unc,
                     color='gray', alpha=0.3, label='Uncertainty')
    plt.title(title)
    plt.xlabel("Sample Index")
    plt.ylabel(title)
    plt.legend()
    plt.show()

# =============================================================================
# Section 3: File Paths & Parameters
# =============================================================================

train_curve_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/6th Hard + soft Properties Mix/Bulk 6th Properties 1-1000.xlsx'
train_props_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/6th Hard + soft Properties Mix/6th Hard + soft Properties Mix.xlsx'

# Prediction (experimental) files
pred_file_tungsten = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Tungsten Experiment updated.xlsx'
pred_file_mo       = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Mo Samples Experiment  Updated.xlsx'
pred_file_copper   = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'
pred_file_simu     = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'  # Additional prediction file

depth_factor  = 1
load_factor   = 1
num_segments  = 1
max_depth_use = 300
exponents     = [1, 2]

# =============================================================================
# Section 4: Load and Scale Training Target σₛ
# =============================================================================

df_props       = pd.read_excel(train_props_file)
target_sigma_s = df_props.iloc[:, 2].values.reshape(-1,1)
scaler_y       = StandardScaler().fit(target_sigma_s)
target_scaled  = scaler_y.transform(target_sigma_s).flatten()
print("Training target σₛ shape:", target_scaled.shape)

plt.figure(figsize=(6,4))
plt.hist(target_scaled, bins=20)
plt.title("Scaled Distribution of σₛ")
plt.xlabel("Scaled σₛ")
plt.ylabel("Frequency")
plt.show()

# =============================================================================
# Section 5: Build and Scale Training Feature Matrix
# =============================================================================

data_train = process_file(train_curve_file, depth_factor, load_factor, 'train', num_segments)
fit_train  = fit_polynomial(data_train, exponents)
train_sel  = {k:v for k,v in fit_train.items() if f"Depth_{max_depth_use}_" in k}

by_sample = {}
for k in train_sel:
    idx = int(k.split('_')[-1])
    by_sample.setdefault(idx, []).append(k)

X_train = []
for idx in sorted(by_sample):
    coeffs = []
    for k in sorted(by_sample[idx]):
        coeffs.extend(train_sel[k])
    X_train.append(coeffs)
X_train = np.array(X_train)

scaler_X     = StandardScaler().fit(X_train)
X_train_scaled = scaler_X.transform(X_train)
print("X_train_scaled shape:", X_train_scaled.shape)

# =============================================================================
# Section 6: Build and Scale Prediction Feature Matrices
# =============================================================================

pred_files = {
    'tungsten': pred_file_tungsten,
    'mo'      : pred_file_mo,
    'copper'  : pred_file_copper,
    'simu'    : pred_file_simu
}

X_preds_scaled = {}
for name, path in pred_files.items():
    Xp = process_prediction_file(
        path, name, depth_factor, load_factor,
        num_segments, max_depth_use, exponents
    )
    Xp_scaled = scaler_X.transform(Xp)
    print(f"{name} X_pred_scaled shape:", Xp_scaled.shape)
    X_preds_scaled[name] = Xp_scaled

# =============================================================================
# Section 7: Train DKL on Scaled σₛ and Predict
# =============================================================================

predictions = {}
for name, Xp_s in X_preds_scaled.items():
    means, unc = train_and_predict(
        target_scaled, X_train_scaled, Xp_s,
        f"σₛ - {name.capitalize()}", num_segments, exponents,
        scaler_y=scaler_y
    )
    predictions[name] = (means, unc)

# =============================================================================
# Section 8: Plot σₛ Predictions
# =============================================================================

for name, (means, unc) in predictions.items():
    plot_predictions(means, unc, f"Predicted σₛ - {name.capitalize()}")
```

```python
# =============================================================================
# Section 1: Install and Import Packages
# =============================================================================

!pip install -q --upgrade git+https://github.com/ziatdinovmax/gpax

import os
import pandas as pd
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

import gpax
import jax.numpy as jnp

# ─────────────────────────────────────────────────────────────────────────────
# PATCH GPAX’s split_in_batches as soon as gpax is imported
# ─────────────────────────────────────────────────────────────────────────────

def split_in_batches_fixed(X_new, batch_size, dim=0):
    """
    Splits array `X_new` into batches along specified dimension `dim`.
    Fixes an UnboundLocalError in the original gpax implementation.
    """
    num_batches = X_new.shape[dim] // batch_size
    X_split = []
    i = -1  # Initialize i so we never reference it before assignment
    for i in range(num_batches):
        if dim == 0:
            X_i = X_new[i*batch_size : (i+1)*batch_size]
        else:
            X_i = X_new[:, i*batch_size : (i+1)*batch_size]
        X_split.append(X_i)

    # Handle leftover (if any)
    # Ensure i is defined even if num_batches is 0
    if num_batches == 0:
        i = -1
    if dim == 0:
        X_i = X_new[(i+1)*batch_size :]
    else:
        X_i = X_new[:, (i+1)*batch_size :]

    if X_i.shape[dim] > 0:
        X_split.append(X_i)

    return X_split

# Monkey‐patch gpax.utils immediately:
gpax.utils.split_in_batches = split_in_batches_fixed

# =============================================================================
# Now you can proceed with your code: plotting style, helper functions, etc.
# =============================================================================

# Set a plotting style based on available styles
available_styles = plt.style.available
if 'seaborn' in available_styles:
    plt.style.use('seaborn')
else:
    plt.style.use('default')

# =============================================================================
# Section 2: Define Helper Functions (process_file, polynomial, fit_polynomial, etc.)
# =============================================================================

def process_file(file_name, depth_factor, load_factor, prefix, num_segments=5):
    df = pd.read_excel(file_name, header=None)
    depths = df.iloc[:, ::2].apply(pd.to_numeric, errors='coerce')
    loads  = df.iloc[:, 1::2].apply(pd.to_numeric, errors='coerce')
    depths *= depth_factor
    loads  *= load_factor

    data_dict = {}
    for i in range(depths.shape[1]):
        depth_col = depths.iloc[:, i].dropna()
        load_col  = loads.iloc[:, i].dropna()
        for max_depth in range(200, 1900, 100):
            mask = (depth_col >= 0) & (depth_col <= max_depth)
            depth_selected = depth_col[mask]
            load_selected  = load_col[mask]
            if not depth_selected.empty:
                segment_length = max_depth / num_segments
                for segment in range(1, num_segments+1):
                    segment_mask = (
                        (depth_selected >= (segment-1)*segment_length) &
                        (depth_selected <  segment*segment_length)
                    )
                    depth_segment = depth_selected[segment_mask]
                    load_segment  = load_selected[segment_mask]
                    if not depth_segment.empty:
                        depth_key = f"{prefix}_Depth_{max_depth}_Segment_{segment}_{i}"
                        load_key  = f"{prefix}_Load_{max_depth}_Segment_{segment}_{i}"
                        data_dict[depth_key] = depth_segment.reset_index(drop=True)
                        data_dict[load_key]  = load_segment.reset_index(drop=True)
    return data_dict

def polynomial(x, *coeffs, exponents):
    return sum(c * x**exp for c, exp in zip(coeffs, exponents))

def fit_polynomial(data_dict, exponents):
    """
    Fits a polynomial (with provided exponents) to each segmented depth curve.
    Returns a dictionary mapping each 'Depth' key to its fitted coefficients.
    Ensures depth and load data have the same length for curve fitting.
    """
    results_dict = {}
    for key in data_dict:
        if 'Depth' in key:
            depth_key = key
            load_key  = key.replace('Depth', 'Load')

            # Align depth and load series based on index after dropping NaNs
            aligned_df = pd.DataFrame({'Depth': data_dict[depth_key], 'Load': data_dict[load_key]}).dropna()

            if not aligned_df.empty:
                X = aligned_df['Depth'].values.flatten()
                y = aligned_df['Load'].values.flatten()

                try:
                    popt, _ = curve_fit(
                        lambda x, *params: polynomial(x, *params, exponents=exponents),
                        X, y, p0=[1]*len(exponents)
                    )
                    results_dict[depth_key] = popt
                except RuntimeError:
                    print(f"Fit failed for dataset {key}")
            else:
                print(f"No valid data points for fitting in dataset {key}")

    return results_dict

def get_dkl_func(targets, X, num_segments, exponents):
    rng_key, rng_key_predict = gpax.utils.get_keys()
    input_dim = len(exponents) * num_segments
    dkl = gpax.viDKL(input_dim, 1, 'RBF')
    print(f"Fitting DKL with input_dim={input_dim}, targets_shape={targets.shape}, X_shape={X.shape}")
    dkl.fit(rng_key, X, targets)
    return dkl, rng_key, rng_key_predict

def process_prediction_file(pred_file, prefix, depth_factor, load_factor, num_segments, max_depth_use, exponents):
    data_pred = process_file(pred_file, depth_factor, load_factor, prefix=prefix, num_segments=num_segments)
    fit_results_pred = fit_polynomial(data_pred, exponents)

    pred_keys = {key: val for key, val in fit_results_pred.items()
                 if f"Depth_{max_depth_use}_" in key}
    segment_keys_by_sample_pred = {}
    for key in pred_keys:
        sample_index = int(key.split('_')[-1])
        segment_keys_by_sample_pred.setdefault(sample_index, []).append(key)

    X_pred = []
    for sample_index in sorted(segment_keys_by_sample_pred.keys()):
        combined_params = []
        for key in sorted(segment_keys_by_sample_pred[sample_index]):
            combined_params.extend(pred_keys[key])
        X_pred.append(combined_params)
    return np.array(X_pred)

def train_and_predict(target, X_train, X_pred, property_name, num_segments, exponents):
    """
    Now that gpax.utils.split_in_batches is already patched at import,
    we don’t need to re-apply the patch here. Simply train & predict.
    """
    dkl, rng_key, rng_key_predict = get_dkl_func(target, X_train, num_segments, exponents)
    predictions = dkl.predict_in_batches(rng_key_predict, X_pred, batch_size=3)
    print(f"Predictions for {property_name}:")
    print(predictions)
    return predictions

def plot_predictions(pred_array, title):
    means, uncert = pred_array
    plt.figure(figsize=(6,4))
    plt.plot(means, 'o-', label='Predicted Mean')
    plt.fill_between(
        np.arange(len(means)),
        means - uncert,
        means + uncert,
        color='gray', alpha=0.3,
        label='Uncertainty'
    )
    plt.title(title)
    plt.xlabel("Sample Index")
    plt.ylabel("Predicted E")
    plt.legend()
    plt.show()

# =============================================================================
# Section 3: File Names & Parameters
# =============================================================================
train_curve_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 dis and load 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'
train_props_file = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/1400 E sigma_y sigma_s 1st hard 1-300 2nd hard 1-200 soft 1-400 5th 1-500.xlsx'

# Prediction (experimental) files
pred_file_tungsten = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/Experiement Data From Different Thickness Copper/Berkovich/1st test/for Correction Test/Top Corner correted spherical Test 1.xlsx'
pred_file_mo       = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/Mo Samples Experiment  Updated.xlsx'
pred_file_copper   = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/ML Excel Spherical Bulk/Experiment from Vivek ML model W Mo Cu/COPPER NEW Experiment Updated.xlsx'
pred_file_simu     = '/content/drive/MyDrive/2024 UTk Research/2024 Fall Thin Film/Experiement Data From Different Thickness Copper/Berkovich/1st test/for Correction Test/Top Corner correted spherical Test 1.xlsx'  # Additional prediction file

depth_factor  = 1
load_factor   = 1
num_segments  = 1
max_depth_use = 300
exponents     = [1, 2]

# =============================================================================
# Section 4: Load & Plot Training Targets
# =============================================================================
df_props = pd.read_excel(train_props_file)
target_E = df_props.iloc[:, 0].values  # Column 0 = E
print("Training target shape for E:", target_E.shape)

plt.figure(figsize=(6,4))
plt.hist(target_E, bins=20, color='skyblue')
plt.title("Distribution of E (Young's Modulus)")
plt.xlabel("E")
plt.ylabel("Frequency")
plt.show()

# =============================================================================
# Section 5: Process Training Curves → X_train
# =============================================================================
data_train = process_file(train_curve_file, depth_factor, load_factor, prefix='train', num_segments=num_segments)
fit_results_train = fit_polynomial(data_train, exponents)

train_keys = {key: val for key, val in fit_results_train.items()
              if f"Depth_{max_depth_use}_" in key}

segment_keys_by_sample = {}
for key in train_keys:
    sample_index = int(key.split('_')[-1])
    segment_keys_by_sample.setdefault(sample_index, []).append(key)

X_train = []
for sample_index in sorted(segment_keys_by_sample.keys()):
    combined_params = []
    for key in sorted(segment_keys_by_sample[sample_index]):
        combined_params.extend(train_keys[key])
    X_train.append(combined_params)
X_train = np.array(X_train)
print("Training feature matrix shape:", X_train.shape)

# Optional sanity‐check plot (one sample’s depth vs. load + fit)
sample_key = list(train_keys.keys())[0]
depth_series = data_train[sample_key]
load_series  = data_train[sample_key.replace('Depth', 'Load')]
coeffs_sample = train_keys[sample_key]

x_vals = np.linspace(depth_series.min(), depth_series.max(), 100)
y_fit = polynomial(x_vals, *coeffs_sample, exponents=exponents)

plt.figure(figsize=(6,4))
plt.plot(depth_series, load_series, 'o', label='Data')
plt.plot(x_vals, y_fit, '-', label='Polynomial Fit')
plt.title(f"Sample Curve and Fit\nKey: {sample_key}")
plt.xlabel("Depth")
plt.ylabel("Load")
plt.legend()
plt.show()

# =============================================================================
# Section 6: Process Prediction Files → X_pred_*
# =============================================================================
X_pred_tungsten = process_prediction_file(
    pred_file_tungsten, prefix='tungsten',
    depth_factor=depth_factor, load_factor=load_factor,
    num_segments=num_segments, max_depth_use=max_depth_use,
    exponents=exponents
)
print("Tungsten prediction feature matrix shape:", X_pred_tungsten.shape)

X_pred_mo = process_prediction_file(
    pred_file_mo, prefix='mo',
    depth_factor=depth_factor, load_factor=load_factor,
    num_segments=num_segments, max_depth_use=max_depth_use,
    exponents=exponents
)
print("Mo prediction feature matrix shape:", X_pred_mo.shape)

X_pred_copper = process_prediction_file(
    pred_file_copper, prefix='copper',
    depth_factor=depth_factor, load_factor=load_factor,
    num_segments=num_segments, max_depth_use=max_depth_use,
    exponents=exponents
)
print("Copper prediction feature matrix shape:", X_pred_copper.shape)

X_pred_simu = process_prediction_file(
    pred_file_simu, prefix='simu',
    depth_factor=depth_factor, load_factor=load_factor,
    num_segments=num_segments, max_depth_use=max_depth_use,
    exponents=exponents
)
print("Simulation prediction feature matrix shape:", X_pred_simu.shape)

# =============================================================================
# Section 7: Train DKL & Predict
# =============================================================================
pred_E_tungsten = train_and_predict(
    target_E, X_train, X_pred_tungsten,
    "E (Young's Modulus) - Tungsten",
    num_segments, exponents
)

pred_E_mo = train_and_predict(
    target_E, X_train, X_pred_mo,
    "E (Young's Modulus) - Mo",
    num_segments, exponents
)

pred_E_copper = train_and_predict(
    target_E, X_train, X_pred_copper,
    "E (Young's Modulus) - Copper",
    num_segments, exponents
)

pred_E_simu = train_and_predict(
    target_E, X_train, X_pred_simu,
    "E (Young's Modulus) - Simulation",
    num_segments, exponents
)

# =============================================================================
# Section 8: Plot Predictions
# =============================================================================
plot_predictions(pred_E_tungsten, "Predicted E (Young's Modulus) - Tungsten")
plot_predictions(pred_E_mo,       "Predicted E (Young's Modulus) - Mo")
plot_predictions(pred_E_copper,   "Predicted E (Young's Modulus) - Copper")
plot_predictions(pred_E_simu,     "Predicted E (Young's Modulus) - Simulation")
```
